```python
import numpy as np
import pandas as pd
import folium
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping

# Load the flood dataset
flood_df = pd.read_csv('/content/flood_prediction_dataset.csv')

# Load the country and city datasets
country_df = pd.read_csv('/content/countries.csv')
city_df = pd.read_csv('/content/cities.csv')

# Combine country and city datasets
country_city_df = pd.merge(
    city_df[['city_name', 'country', 'latitude', 'longitude']],
    country_df[['country', 'capital_lat', 'capital_lng']],
    on='country',
    how='left'
)

# Randomly assign a country to each row in flood_df from the country_city_df's 'country' column
countries = country_city_df['country'].unique()
np.random.seed(42)  # For reproducibility
flood_df['country'] = np.random.choice(countries, size=len(flood_df))

# Merge the flood dataset with the combined country_city_df based on 'country'
flood_df_combined = pd.merge(flood_df, country_city_df, on='country', how='left')

# Fill missing latitude/longitude with the capital city's lat/lng as a fallback
flood_df_combined['latitude'].fillna(flood_df_combined['capital_lat'], inplace=True)
flood_df_combined['longitude'].fillna(flood_df_combined['capital_lng'], inplace=True)

# Drop unnecessary columns
flood_df_combined.drop(columns=['capital_lat', 'capital_lng'], inplace=True)

# Data Preprocessing
scaler = StandardScaler()
X = flood_df_combined.drop(columns=['FloodProbability', 'country', 'city_name'])  # All features except the target and non-numeric data
y = flood_df_combined['FloodProbability']  # Target variable

X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model building
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='linear')  # Regression output (flood probability)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_test, Y_test), callbacks=[early_stopping])

# Evaluate the model
predictions = model.predict(X_test)
mse = np.mean((Y_test - predictions.squeeze()) ** 2)
r2 = 1 - (np.sum((Y_test - predictions.squeeze()) ** 2) / np.sum((Y_test - np.mean(Y_test)) ** 2))

print(f"Model Performance - MSE: {mse}, R2: {r2}")

# Save the model
model.save('flood_prediction_model.h5')

# Adding predicted flood probability to the DataFrame for visualization
flood_df_combined['PredictedFloodProbability'] = model.predict(X_scaled)

# Filter high-risk areas (flood probability > 0.5)
high_risk_areas = flood_df_combined[flood_df_combined['PredictedFloodProbability'] > 0.5]

# Extract latitude, longitude, and probabilities for high-risk areas
latitudes = high_risk_areas['latitude'].tolist()
longitudes = high_risk_areas['longitude'].tolist()
probabilities = high_risk_areas['PredictedFloodProbability'].round(2).tolist()

# Generate an interactive map using Folium
if len(latitudes) > 0 and len(longitudes) > 0:
    # Create a map centered around the first high-risk area
    m = folium.Map(location=[latitudes[0], longitudes[0]], zoom_start=5)

    # Add markers
    for lat, lng, prob in zip(latitudes, longitudes, probabilities):
        folium.Marker(
            location=[lat, lng],
            popup=f'Flood-prone area with probability: {prob}',
            icon=folium.Icon(color='red')
        ).add_to(m)

    # Display map (in Jupyter Notebooks) or save it
    m.save("flood_prone_areas_map.html")  # Save as HTML if needed or use m to display in Jupyter
    print("Map of high-risk flood areas saved as 'flood_prone_areas_map.html'.")
else:
    print("No high-risk areas detected.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import StandardScaler

# Load the saved model
model = tf.keras.models.load_model('flood_prediction_model.h5')



# Load and preprocess your data
flood_df = pd.read_csv('/content/flood_prediction_dataset.csv')

# Load the country and city datasets
country_df = pd.read_csv('/content/countries.csv')
city_df = pd.read_csv('/content/cities.csv')

# Combine country and city datasets
country_city_df = pd.merge(
    city_df[['city_name', 'country', 'latitude', 'longitude']],
    country_df[['country', 'capital_lat', 'capital_lng']],
    on='country',
    how='left'
)

# Randomly assign a country to each row in flood_df from the country_city_df's 'country' column
countries = country_city_df['country'].unique()
np.random.seed(42)  # For reproducibility
flood_df['country'] = np.random.choice(countries, size=len(flood_df))

# Merge the flood dataset with the combined country_city_df based on 'country'
flood_df_combined = pd.merge(flood_df, country_city_df, on='country', how='left')

# Fill missing latitude/longitude with the capital city's lat/lng as a fallback
flood_df_combined['latitude'].fillna(flood_df_combined['capital_lat'], inplace=True)
flood_df_combined['longitude'].fillna(flood_df_combined['capital_lng'], inplace=True)

# Drop unnecessary columns
flood_df_combined.drop(columns=['capital_lat', 'capital_lng'], inplace=True)

# Data Preprocessing
X = flood_df_combined.drop(columns=['FloodProbability', 'country', 'city_name'])  # All features except the target and non-numeric data
X_scaled = scaler.transform(X)  # Scale the features using the saved scaler

# Predict flood probabilities
flood_df_combined['PredictedFloodProbability'] = model.predict(X_scaled).flatten()

# Filter high-risk areas (flood probability > 0.5)
high_risk_areas = flood_df_combined[flood_df_combined['PredictedFloodProbability'] > 0.5]
```

```python
# Scatter plot for high-risk areas
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', hue='PredictedFloodProbability', size='PredictedFloodProbability', sizes=(20, 200), data=high_risk_areas, palette='Reds', legend=None)
plt.title('High-Risk Flood Areas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Predicted Flood Probability')
plt.show()
```
